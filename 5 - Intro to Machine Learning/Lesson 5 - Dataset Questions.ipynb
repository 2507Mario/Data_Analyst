```python
#!/usr/bin/python

""" 
    Starter code for exploring the Enron dataset (emails + finances);
    loads up the dataset (pickled dict of dicts).

    The dataset has the form:
    enron_data["LASTNAME FIRSTNAME MIDDLEINITIAL"] = { features_dict }

    {features_dict} is a dictionary of features associated with that person.
    You should explore features_dict as part of the mini-project,
    but here's an example to get you started:

    enron_data["SKILLING JEFFREY K"]["bonus"] = 5600000
    
"""

import pickle

enron_data = pickle.load(open("../final_project/final_project_dataset.pkl", "r"))

#print enron_data.keys()

# Count 'Em Up
print "There are", len(enron_data), "persons in the dataset"

# For each person, how many features are available?
print "There are", len(enron_data["SKILLING JEFFREY K"]), "features per person"

# How many POIs are there in the E+F dataset?
print "There are", len(dict((key, value) for key, 
                                  value in enron_data.items()
                            if value["poi"] == True)), "POIs in the E+F dataset"

# How many POI’s were there total?
pois = open("../final_project/poi_names.txt", "r")
pois.readline() # do not the link line
pois.readline() # do not the blank line
pois_count = 0
for pois in pois:
	pois_count += 1
print "There are", pois_count, "total POIs"

# What is the total value of the stock belonging to James Prentice?
print "The total value of James Prentice´s stock is",enron_data["PRENTICE JAMES"]["total_stock_value"]

# How many email messages do we have from Wesley Colwell to persons of interest?
print "There are",enron_data["COLWELL WESLEY"]["from_this_person_to_poi"], "email messages from W. C. to poi"

# What’s the value of stock options exercised by Jeffrey K Skilling?
print "The value of stock options exercised by Jeffrey Skilling is",enron_data["SKILLING JEFFREY K"]["exercised_stock_options"]

# Of these three individuals (Lay, Skilling and Fastow), 
# who took home the most money (largest value of “total_payments” feature)? 
# How much money did that person get?
most_paid = ''
highest_payment = 0

for key in ('LAY KENNETH L', 'FASTOW ANDREW S', 'SKILLING JEFFREY K'):
	if enron_data[key]['total_payments'] > highest_payment:
		highest_payment = enron_data[key]['total_payments']
		most_paid = key

print most_paid, "took the highest payment,",highest_payment

# For nearly every person in the dataset, not every feature has a value. 
# How is it denoted when a feature doesn’t have a well-defined value?
#print(enron_data['SKILLING JEFFREY K'])

# How many folks in this dataset have a quantified salary?
quantified_salary = len(dict((key, value) for key, value in enron_data.items() if value["salary"] != 'NaN'))
print quantified_salary, "people have a quantified salary"

# What about a known email address?
email_address = len(dict((key, value) for key, value in enron_data.items() if value["email_address"] != 'NaN'))
print email_address, "people have an email address"

# How many people in the E+F dataset (as it currently exists) have “NaN” for their total payments? 
#What percentage of people in the dataset as a whole is this?
Nan_total_payments = len(dict((key, value) for key, value in enron_data.items() if value["total_payments"] == 'NaN'))
print Nan_total_payments, "people have NaN total payments"

total_payments = len(dict((key, value) for key, value in enron_data.items() if value["total_payments"]))
print total_payments, "are the total number of people"

from __future__ import division
Percent = float(Nan_total_payments) / total_payments * 100
print "That is", Percent, "%"

#How many POIs in the E+F dataset have “NaN” for their total payments? 
#What percentage of POI’s as a whole is this?
who_are_pois = dict((key, value) for key, value in enron_data.items() if value["poi"] == True)
who_are_pois_len = len(who_are_pois)
print who_are_pois_len, "people are pois"

Nan_tot_pay_pois = len(dict((key, value) for key, value in who_are_pois.items() if value["total_payments"] == 'NaN'))
print Nan_tot_pay_pois, "pois have NaN total payments"

from __future__ import division
Percent = float(Nan_tot_pay_pois) /  who_are_pois_len* 100
print "That is", Percent, "%"
```
